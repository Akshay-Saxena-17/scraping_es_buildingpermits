```python
import requests
import os
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Website
url = "https://apps.fomento.gob.es/BoletinOnline/?nivel=2&orden=09000000"

#get all .XLS hrefs
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
links = [link.get('href') for link in soup.find_all('a') if '.XLS' in link.get('href')]
```

```python
#We'll make 3 materfiles for each of the categories.
#Category 1: New Residential units and average area built along with major renovation projects (that require permits)
dfmaster1 = pd.DataFrame()

#Iterating through the links collected
for link in links:

#     Only want to see province level category 1 files
    if link[9] == '3' and link[12] == '1':
#         Want to see what link we are on
        print(link)
        
#         Collecting the data
        r=requests.get("https://apps.fomento.gob.es/BoletinOnline/" + link, allow_redirects = True)
        
#         Naming the file
        filename = link[6:14]+'.xls'
        
#         Putting it into the file
        with open (os.path.join("./",filename),'wb') as f:
            f.write(r.content)
        
#         Reading the XLS file into a dataframe
        dftemp = pd.read_excel('./'+filename)

#         Filling the missing years
        dftemp['Unnamed: 0'].fillna(method='ffill', inplace = True)
    
#         Adding the cpro to join with other data sets
        dftemp['cpro'] = link[10:12]
    
#         Adding the typeflag (category)
        dftemp['typeflag'] = link[12:13]
    
#         Just want to add the month wise data
        dftemp2 = dftemp[~dftemp['Unnamed: 0'].isnull() & ~dftemp['Unnamed: 2'].isnull()]
    
        dfmaster = pd.concat([dfmaster,dftemp2])

    else: 
        continue
```
